# Decision Tree Classifier

```python
dt['Gender']=dt['Gender'].astype('category')
dt['OverTime']=dt['OverTime'].astype('category')
```

```python
gender_dummy=pd.get_dummies(dt['Gender'])
ot_dummy=pd.get_dummies(dt['OverTime'])
```

```python
dt=pd.concat((dt,gender_dummy, ot_dummy),axis=1)
```

```python
dt=dt.drop(['Gender','OverTime'],axis=1)
```

```python
dt.head()
```

```python
from sklearn.model_selection import train_test_split
X=dt.drop(['Attrition','BusinessTravel','Department','EducationField','JobRole','MaritalStatus'],axis=1)
y=dt['Attrition']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)
#gender and overtime
```

```python
from sklearn.tree import DecisionTreeClassifier
mdl=DecisionTreeClassifier(max_depth=10,min_samples_leaf=10,min_samples_split=20)
mdl=mdl.fit(X_train,y_train)
```

```python
mdl.get_params()
```

```python
X_test
```

```python
predict=mdl.predict(X_test)
predict
```

```python
mdl.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predict)
```
